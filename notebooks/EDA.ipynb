```python
plt.hist(df['recency'], bins=20)
plt.xlabel('Recency')
plt.ylabel('Frequency')
plt.title('Distribution of Recency')
plt.show()  
```

```python
plt.hist(df['total_payment'], bins=100)
plt.xlabel('Total Payment')
plt.ylabel('Frequency')
plt.title('Distribution of Total Payment')
plt.show()  
```

```python
df['churn_status'].value_counts().plot(kind='bar')
plt.xlabel('Churn Status')
plt.ylabel('Count')
plt.title('Distribution of Churn Status')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Maan lijiye aapka dataframe 'df' hai aur column 'Total Payment'
# 1. Log transformation apply karna
df['Log_Total_Payment'] = np.log1p(df['total_payment'])

# 2. Result check karne ke liye histogram plot karein
plt.figure(figsize=(10, 5))

# Purana skewed data
plt.subplot(1, 2, 1)
sns.histplot(df['total_payment'], bins=50)
plt.title('Original (Skewed) Data')

# Naya transformed data
plt.subplot(1, 2, 2)
sns.histplot(df['Log_Total_Payment'], bins=50)
plt.title('After Log Transformation')

plt.show()
```
